### 4.Insights

```python
def limpar_valor(valor_texto):
    if isinstance(valor_texto, str):
        # 1.Remove o euro e espaços
        valor_texto = valor_texto.replace('€', '').replace(' ', '')

        # 2.Se for Milhão (M)
        if 'M' in valor_texto:
            valor_texto = valor_texto.replace('M', '')
            return float(valor_texto) * 1000000

        # 3.Se for Mil (K)
        elif 'K' in valor_texto:
            valor_texto = valor_texto.replace('K', '')
            return float(valor_texto) * 1000

    return 0.0 #Caso o valor seja vazio ou 0


#Aplicando na coluna Value
df['Value_Num'] = df['Value'].apply(limpar_valor)

#Aplicando na coluna Wage
df['Wage_Num'] = df['Wage'].apply(limpar_valor)

#Conferindo o resultado
print(df[['Name', 'Value', 'Value_Num']].head())
```

```python
value_players = df[['Name', 'Age', 'Value_Num','Wage_Num']]
value_players.sort_values('Value_Num', ascending =False)
```

```python
value_players.sort_values('Wage_Num', ascending =False)
```
